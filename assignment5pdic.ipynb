```python
# Q1. Repeat a tuple three times using * operator

t = (1, 2, 3)
print(t * 3)
```

```python
# Q2. Join three separate tuples into one new tuple

t1 = (1, 2)
t2 = (3, 4)
t3 = (5, 6)

new_tuple = t1 + t2 + t3
print(new_tuple)
```

```python
# Q3. Check whether a specific element exists inside a tuple

t = (10, 20, 30, 40)
element = 30

if element in t:
    print("Element exists")
else:
    print("Element does not exist")
```

```python
# Q4. Calculate total, highest value and lowest value without sum(), max(), min()

t = (12, 45, 7, 89, 23)

total = 0
highest = t[0]
lowest = t[0]

for i in t:
    total += i

    if i > highest:
        highest = i

    if i < lowest:
        lowest = i

print("Total =", total)
print("Highest =", highest)
print("Lowest =", lowest)
```

```python
# Q5. Keep only values greater than 10

n = (3, 14, 7, 22, 9, 41, 18, 5)

result = ()

for i in n:
    if i > 10:
        result += (i,)

print(result)
```

```python
# Q6. Determine how many elements are in a set without len()

s = {"cat", "dog", "bird", "fish"}

count = 0

for i in s:
    count += 1

print("Number of elements =", count)
```

```python
# Q7. Combine two sets into one

s1 = {1, 2, 3}
s2 = {3, 4, 5}

result = s1.union(s2)

print(result)
```

```python
# Q8. Find all elements common to both sets

s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

result = s1.intersection(s2)

print(result)
```

```python
# Q9. Find all elements that are in either set but not both

s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

result = s1.symmetric_difference(s2)

print(result)
```
